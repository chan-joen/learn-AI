```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
```

```python
x= diabetes.data[:,2]
y=diabetes.target
```

```python
w=1.0
b=1.0
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
y_hat=w*x+b
plt.plot(x,y_hat)
plt.show()
```

```python
def forpass(x):
  y_hat=x*w+b
  return y_hat
```

```python
def backprop(x,err):
  w_grad=x*err
  b_grad=1*err
  return w_grad,b_grad
```

```python
def fit(x,y,epochs=100):
  global w
  global b
  for i in range(epochs):
    for x_i,y_i in zip(x,y):
      y_hat = forpass(x_i)
      err = -(y_i-y_hat)
      w_grad,b_grad=backprop(x_i,err)
      w-=w_grad
      b-=b_grad
```

```python
fit(x,y)
```

```python
plt.scatter(x,y)
y_hat=w*x+b
plt.plot(x,y_hat)
plt.show()
```
